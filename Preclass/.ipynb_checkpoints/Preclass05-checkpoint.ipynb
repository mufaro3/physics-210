## *5.2 Simple Plots*

Notice that, similar to the way we have often used `import numpy as np`, we typically use
`import matplitlib.pyplot as plt`. 

**A simple 2d plot:** Let’s get started with a simple plot. 

````python
# Run me!

import matplotlib.pyplot as plt

xdata = [1, 2, 3] # xdata as a list
ydata = [3, 4, 6] # ydata as a list
plt.plot(xdata, ydata) # plot y vs x
plt.show()
````

**Matplotlib can plot lists or arrays.**

````python
# Run me to see an example of plotting arrays instead of lists!

import numpy as np
import matplotlib.pyplot as plt

xdata = np.array([1, 2, 3]) # xdata as an array
ydata = np.array([3, 4, 6]) # ydata as an array
plt.plot(xdata, ydata) # plot y vs x
plt.show()
````

**Adding markers for the data points, changing colour, adding labels/titles and modifying axis ranges manually:**: A third, optional argument in `plot()` is related to point/line/color style. The options are:

* `rgbcymkw` : red, green, blue, cyan, yellow, magenta, black, white
* "`-`" line "`--`" dashed line "`-.`" dash-dot line ":" dotted line
* "`o`" circle "`+`" plus "`x`" x-symbol "`s`" square "`*`" star
* "`^`" up-triangle "`v`" down-triangle "`D`" diamond
* Default "`b-`" is a solid blue line with *no points*
* "`ro--`" is red circles connected by red dashed lines

For example, the following script uses red circles for the data points and joins them with a solid red line (`"ro-"`). The example also shows you how you can alter the x and y ranges and add titles/labels:

````python
# Run me!

import matplotlib.pyplot as plt

xdata = [1, 2, 3]
ydata = [3, 4, 6]

plt.plot(xdata, ydata, "ro-")
plt.title("This is the Title")
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.xlim(0.0, 4.0)
plt.ylim(2.5, 6.5)
plt.show()
````

## *5.3 Adding error bars*
Instead of just being an extra argument in `plt.plot()`, a plot with error bars requires a different function `plt.errorbar()`.

````python
# Run me!
import matplotlib.pyplot as plt

xdata = [1, 2, 3]
ydata = [3, 4, 6]
yerrs = [0.5, 1, 2]
plt.errorbar(xdata, ydata, yerrs)
plt.show()
````

**Important (and annoying)!** In contrast to how line styles are specified in `plt.plot()`, in `plt.errorbar()` we need to use the `fmt` (format) keyword when specifying line styles. Both graph types use the same options. 

In out example below,  we just make them red points: `fmt="ro"`:

````python
# Run me!

import matplotlib.pyplot as plt

xdata = [1, 2, 3]
ydata = [3, 4, 6]
yerrs = [0.5, 1, 2]
plt.errorbar(xdata, ydata, yerrs, fmt="ro")
plt.show()
````

## *5.6 Multiple Plots `plt.subplots()` and `plt.subplot()`*

There are two similarly named functions in Matplotlib to plot multiple graphs on the same canvas: `subplots()` and `subplot()` The `subplots()` function is considered a more modern approach than `subplot()`.

### 5.6.1 Multiple plots function 1: `plt.subplots()`

The example below uses `fig, ax = plt.subplots(2, 1)` to greate a figure-type object `fig`, with an array of "axis" objects `ax`. In this specific case there are two of these axes objects, where the first one is `ax[0]` and the second is `ax[1]`. So you can see that everything we want to do with the first plot uses `ax[0].something` and for the second plot `ax[1].something`.

**Important!** When using `plt.subplots()`, we add axis labels using notation of the form `ax[0].set_xlabel()` instead of `ax[0].xlabel()` like we did with our previous matplotlib graphs.

````python
# Run me!

import numpy as np
import matplotlib.pyplot as plt

# Create data sets
t = np.linspace(0, 1.9, 20) # 20 evenly spaced numbers from 0 to 1.9
y1 = np.exp(-0.6 * t) * np.cos(6.5 * t + 0.5)
y2 = np.cos(2 * np.pi * t)

fig, ax = plt.subplots(2, 1) # Can also be expressed as plt.subplots(2)

ax[0].plot(t, y1, 'bo', t, y1, 'k')
ax[0].set_xlabel("t (s)") 
ax[0].set_ylabel("y1 (m)")

ax[1].plot(t, y2, 'r--')
ax[1].set_xlabel("t (s)")
ax[1].set_ylabel("y2 (m)")

plt.tight_layout() # Adjusts spacing between subplots for better layout
plt.show()
````

### 5.6 Multiple plots function 2: `plt.subplot()`

The `plt.subplot()` function has some significant differences from `plt.subplots()`. 

Instead of specifying the overall configuration of the plots in a `fig, ax = plt.subplots(2, 1)` type of line, we instead specify the configration of the subplots at the same time we indicate which subplot we are working on. The line `plt.subplot(numrows, numcols, subplotnum)` tells us that our configuration will have `numrows` rows and `numcols` columns. And then `subplotnum` sets the current subplot that will be interacted with with subsequent matplotlib commands.

In contrast to `plt.subplots()`, when we use `plt.subplot()` all of the functions used to interact with a given subplot are the same as the ones we previously learned for other matplotlib objects. 

````python
# Run me!

import numpy as np
import matplotlib.pyplot as plt

# Create data sets
t = np.linspace(0, 1.9, 20)  # 20 evenly spaced numbers from 0 to 1.9
y1 = np.exp(-0.6 * t) * np.cos(6.5 * t + 0.5)
y2 = np.cos(2 * np.pi * t)

# First subplot
plt.subplot(2, 1, 1)
plt.plot(t, y1, 'bo', t, y1, 'k')
plt.xlabel("t (s)")
plt.ylabel("y1 (m)")

# Second subplot
plt.subplot(2, 1, 2)
plt.plot(t, y2, 'r--')
plt.xlabel("t (s)")
plt.ylabel("y2 (m)")

plt.tight_layout()
plt.show()
````

## *5.7 Histograms*
The function`plt.hist()` creates a histogram of values in an `array` or `arrays`. 

The following script will first create a normal distribution of random values and then makes a histogram. You may wish to lookup (Google, ChatGPT,...) exactly what `np.random.randn(10000)` does.

````python
# Run me to see a basic histogram
# - Run me multiple times to see thr random data set change each time

import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10000) # 10,000 randomly generated values

# Plot a basic histogram
plt.hist(x)
plt.show()
````

Let's introduce some additional arguments in our `plt.hist` call, as well as adding axis labels and making other modifications to the graph. 

Have a look at the following refernce page to learn more about arguments that `plt.hist` can accept:
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html


````python
# Run me!

import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10000) # 10,000 randomly generated values

# Plot a histogram of the values below
plt.hist(x,50,facecolor='g',alpha=0.75, density=True)

plt.xlabel('IQ')
plt.ylabel('Probability')

# See below for a discussion of of plt.text()
plt.text(45, .027, f"$\mu={mu}, \sigma={sigma}$", fontsize=16)
plt.axis([40, 160, 0, 0.03]) # Set axis bounds manually
plt.grid(True)
plt.show()
````

Notice that the `plt.text()` method accepts LaTeX syntax text and can also use the string formatting tricks that we've been learning. If you are new to LaTeX, there is a library of common symbols here: http://artofproblemsolving.com/wiki/index.php?title=LaTeX:Symbols 

## *5.8 Scatter Plots*

The example below is a 2-dimensional scatter plot with a bar next to it to indicate the colour scheme. `plt.plot()` can also create scatter plots, but is primarily used for creating line plots. In contrast, `plt.scatter()` is used when you want to have control over the characteristics of the individual data points, such as size or colo(u)r.

````python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
colors = [10, 20, 30, 40, 50]
sizes = [100, 200, 300, 400, 500]
plt.scatter(x, y, c=colors, s=sizes)  # Custom colors and sizes
plt.colorbar()  # show color scale
plt.show()
````
