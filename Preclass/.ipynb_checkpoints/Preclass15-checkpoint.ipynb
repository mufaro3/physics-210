## *15.1 Introduction*

We are going to learn how to model various types of processes using random numbers:

* **Random processes** - One example is radioactive decay, where it is not possible to know exactly when any particular atom will decay, and
* **Processes that are most easily modeled or approximated using randomness** - Brownian motion is much more practical to model as a random process, but if we knew the positions and velocities of all of the surrounding molecules, we could model it deterministically. 

Let's look at how randomness works in computing as we start learning how to use random inputs to model systems and solve problems computationally, which is most commonly known as a Monte Carlo (MC) simulation or method. At their core Monte Carlo methods involve:

1. Drawing a set of random inputs from a probability distribution.
2. Performing deterministic calculations on these inputs and/or drawing further sets of random inputs to model intermediate behaviours and interactions.
3. Aggregating these results to get an approximation.

## *15.2 Pseudorandom numbers*

Let's generate our first random numbers. When numpy, and similar libraries perform "random" operations, they usually return "pseudorandom" results. Pseudorandom numbers are numbers that are generated in a deterministic manner, using an algorithm, but which appear to be random. As we will see in a moment, when we provide an algorithm that generates pseudorandom numbers the same initial state (typically known as a "seed"), the sequence of pseudorandom numbers will always be the same. This is in contrast to true random numbers, which are not deterministic and cannot be predicted. Importantly, for our purposes, the statistical behvaiour of these pseudorandom numbers is indistiguishable from true random numbers.

If you would like to learn more about how numpy performs these operations, you can read the detailed account of it here: https://numpy.org/doc/stable/reference/random/index.html

Let's create an array of 5 random numbers. Run this cell multiple times to see that the numbers continue to be different each time they are generated.

````python
# np.random.rand returns random floats in the half-open interval [0.0, 1.0).

import numpy as np
np.random.rand(5)
````

Let's briefly look at using a seed to generate these numbers. It is important to recognize that if we do not specify the seed, a default seed is provided to the function using a nondeterministic piece of information from the computing system, such as the current time.

When we specify a seed of `8000` below, we are telling the random number generating algorithm the initial state from which it should generate all of its pseudorandom numbers. 

````python
# Run this cell multiple times and notice the output never changes

np.random.seed(8000) # seed = 8000
np.random.rand(5)
````

## *15.3 Using Monte Carlo methods to approximate the value of π*

Let's look at a common introductory example of using the technique of Monte Carlo methods, which is to approximate the value of π. 

To so this we will

1. Randomly generate points uniformly inside a unit square ($R=1$).
2. Determine how many of these points fall inside a quarter circle inscribed in the square.
   * The area of the quarter circle with $R=1$ is given by
   
   $$A_{\mbox{quarter circle}} = \frac{\pi R^2}{4} = \frac{\pi 1^2}{4} = \frac{\pi}{4}.$$
   
   * The area of the unit square $R \times R$ is
   
   $$A_{\mbox{unit square}} = 1.$$
   <br>
3. The ratio of the points inside the quarter circle ($N_{in}$) to the total number of points ($N_{tot}$) gives

$$\frac{N_{in}}{N_{tot}} = \frac{A_{\mbox{quarter circle}}}{A_{\mbox{unit square}}}=\frac{\pi / 4}{1}=\frac{\pi}{4}.$$

4. Finally, to get π, we need to multiply this ratio $N_{in}/N_{tot}$ by 4:

$$4 \frac{N_{in}}{N_{tot}} = 4\frac{\pi}{4} = \pi.$$

Let's visualize how this will work.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(500)
y = np.random.rand(500)

plt.axes().set_aspect(1)

# Points that fall inside the quarter circle
mask = x**2 + y**2 <= 1
plt.plot(x[mask], y[mask], 'ro')

# Points htat fall outside the quarter circle
mask = x**2 + y**2  >1
plt.plot(x[mask], y[mask], 'bo')

plt.xlabel("x value")
plt.ylabel("y value")
plt.show()
````

Before we modify the above code to do our actual calculation, let's look at some criteria for the random numbers we used for this example. These are summarized by Ayars in his book (https://freecomputerbooks.com/Computational-Physics-with-Python-by-Eric-Ayars.html):

1. They must be **uniformly distributed**. For exmaple, if the points were distributed about a normal curve then we'd see more points in the central area of the figure and get a much less accurate number.
1. They must be **uncorrelated**. For example, if our random numbers were correlated in such a way that a low number was most often followed by a high number, and vice-versa, then the points in the figure would trend towards the upper left and lower right of the figure, skewing our results.
1. We need **a lot of points** for an accurate approximation. The statistical uncertainty from this method, expressed as a percent/fractional error scales approximately as $1/\sqrt{N}$ (where $N$ is our total number of points). 

Let's perform the calculation. 

````python
# Run me to approximate pi using N = 10,000 samples
# - Try running it multiple times to see how the MC values changes each time

import numpy as np

def monte_carlo_pi(N):
    x = np.random.rand(N)
    y = np.random.rand(N)
    d = x**2 + y**2
    
    # N is our total number of Monte Carlo events
    # count is our number that fall within the quarter circle
    count = np.sum(d <= 1)
    return 4*count/N

print( f" MC: {monte_carlo_pi(10000):.4f}, Actual: {np.pi:.4f}")
````

## *15.5 A simple Monte Carlo simulation of radioactive decay*

Below is a relatively simple Monte Carlo simulation for the decay of 1000 atoms of $^{208}$Tl, which has a half-life of 3.053 minutes. The key idea is that our current number of radioactive atoms $N$ decays over time (e.g., it should instead be expressed as $N(t)$). At each time step, we need to use the decay probability `p`, that a given radioactive atom will decay during that time step. How do we determine this probability?

We start with the exponential decay law:

$$N(t) = N_0 e^{-\lambda t}.$$

Then we say that the fraction of atoms that will have decayed between $t$ and $t + \delta t$ can be expressed as:

$$\frac{N(t) - N(t + \delta t)}{N(t)}$$

$$= \frac{N_0 e^{-\lambda t} - N_0 e^{-\lambda (t + \delta t)}}{N_0 e^{-\lambda t}}$$ 

$$= 1 - \frac{e^{-\lambda (t + \delta t)}}{e^{-\lambda t}} $$

$$= 1 - e^{-\lambda \delta t}$$

Importantly, the fraction of atoms that will have decayed between $t$ and $t + \delta t$ is the same as the probability `p` that a given radioactive atom will have decayed during that time step. For example, if 77 of 100 atoms decayed during a time step $\delta t$, we could also say that each individual atom had a 77% chance of decaying during that time step.

Finally, we can use the following relationship between the decay constant $\lambda$ and the half-life $T_{1/2}$

$$ T_{1/2} = \dfrac{\ln(2)}{\lambda},$$ 

to rewrite our probability equation in terms of $T_{1/2}$:

$$ p = 1 - e^{-dt \ln(2) / T_{1/2} } .$$

To recap, at each time step, we need use the decay probability `p` above and $N$ random numbers to determine how many of the $N$ remaining radioactive numbers decayed during that time step.

**Take some time to try to understand this simulation. One of the self-assessment questions will follow-up on this simulation.**

````python
import numpy as np
import matplotlib.pyplot as plt

def radioactive_decay(N, half_life, delta_t, time_steps):
    """Simulate radioactive decay using Monte Carlo method.
    
    Parameters:
    - N: Initial number of radioactive atoms
    - half_life: Half-life in seconds
    - delta_t: Time step for the simulation in seconds
    - time_steps: Total number of time steps to simulate
    
    Returns:
    - A list of the number of radioactive atoms remaining at each time step.
    """
    
    # Probability out of 1 that a given atom will decay during a 
    # time interval delta_t
    p = 1 - np.exp(-delta_t * np.log(2) / half_life)
    
    # Start our list for numbers of atoms remaining after each time step
    atoms_remaining = [N]
    
    # Loop through our total number of time_steps
    for t in range(time_steps):
        
        # Determine how many of the N atoms remaining have decayed 
        # during this time step
        decayed = np.sum(np.random.rand(N) < p)
        
        # Update our total number of radioactive atoms 
        N -= decayed
        
        # Update the list of atoms remaining after each time step
        atoms_remaining.append(N)
    
    # Return the list of atoms remaining after each time step
    return atoms_remaining

# Parameters
N_0 = 1000               # Initial number of atoms
half_life = 3.053 * 60 # Half-life in seconds (3.053 minutes converted to seconds)
delta_t = 1            # Time step: 1 second
time_steps = 500       # Number of time steps

# Run simulation
atoms = radioactive_decay(N_0, half_life, delta_t, time_steps)

### Plot the results of N(t) vs t

times = np.linspace(0, time_steps,time_steps+1) # 500 steps after t = 0

# 'times' is an array so 'predicted' an array of the same shape, corresponding to
# the prediced number of atoms that will be left after each time step
predicted = N_0*np.exp(-times*np.log(2) / half_life)

plt.plot(times, atoms, label="Monte-Carlo")
plt.plot(times, predicted, label="Predicted")
plt.xlabel('Time (seconds)')
plt.ylabel('Remaining Atoms')
plt.title(f'Radioactive Decay Simulation for {280}Tl\n(Half-life = {half_life/60} mins)')
plt.legend()
plt.grid(True)
plt.show()
````
