# Project 1 Distinction Proposal

I'm intending to reach distinction level 2 with this project.

## Physics and Physical Modelling

I believe that this is where my project stands out the most. My project utilizes a variety of physics, including jet propulsion, soft-body collision, and 3-body gravitational mechanics. Mathematically, my project also utilizes multi-objective optimization and some more complex linear algebra in the form of vector rotation. 

## Coding

This is my second-best area of this project. My programming techniques include extensive use of subroutines, structures (both classes and dataclasses), and a mix of commenting and explicit variables (particularly the inclusion of docstrings) for clarity. Additionally, extensive use of animations is used for verification, and this is beyond the basic expectation of static graphs.

## Communication of investigation

I feel as though my communication was particularly successful through my explanation of the physics and final optimization model, and my analyses under each plot were particularly thorough. I spent a lot of time generating detailed animations to show the mechanics of the simulation visually.

## Quality of investigation

I'm certain that my investigation was, overall, fairly solid. My exploration of the speed, angle phase space has a full resolution of $10 \times 10$ points, and I made sure to look fairly deep into individual cases of the simulation, such as the initial example and the optimal case.
